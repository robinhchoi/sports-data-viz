```python
import pandas as pd
import json
from nba_py import player
```

```python
test_df = pd.read_csv('ATL_2017.csv')
test_df = test_df[["player", "salary", "team_rank"]]
test_df["team"] = 'ATL'
test_df
```

```python
# Eastern Conference divisions
atlantic = {'BOS': 'Boston Celtics', 'BRK': 'Brooklyn Nets', 'NYK': 'New York Knicks', 'PHI': 'Philadelphia 76ers', 'TOR': 'Toronto Raptors'}
central = {'CHI': 'Chicago Bulls', 'CLE': 'Cleveland Cavaliers', 'DET': 'Detroit Pistons', 'IND': 'Indiana Pacers', 'MIL': 'Milwaukee Bucks'}
southeast = {'ATL': 'Atlanta Hawks', 'CHO': 'Charlotte Hornets', 'MIA': 'Miami Heat', 'ORL': 'Orlando Magic', 'WAS': 'Washington Wizards'}

# Western Conference divisions
northwest = {'DEN': 'Denver Nuggets', 'MIN': 'Minnesota Timberwolves', 'OKC': 'Oklahoma City Thunder', 'POR': 'Portland Trail Blazers', 'UTA': 'Utah Jazz'}
pacific = {'GSW': 'Golden State Warriors', 'LAC': 'Los Angeles Clippers', 'LAL': 'Los Angeles Lakers', 'PHO': 'Phoenix Suns', 'SAC': 'Sacramento Kings'}
southwest = {'DAL': 'Dallas Mavericks', 'HOU': 'Houston Rockets', 'MEM': 'Memphis Grizzlies', 'NOP': 'New Orleans Pelicans', 'SAS': 'San Antonio Spurs'}

# Conferences lists
east_teams = [atlantic, central, southeast]
west_teams =[northwest, pacific, southwest]

all_teams = [east_teams, west_teams]
```

```python
ex = pd.DataFrame({"player": [], "salary": [], "team_rank": [], "team": []})
ex
a = ex.append(test_df)
a
```

```python
total_df = pd.DataFrame({"player": [], "salary": [], "team_rank": [], "team_abbrev": [], "team_name": []})

for conf in all_teams:
    for div in conf:
        for team in div:
            team_df = pd.read_csv(team + "_2017.csv")
            team_df = team_df[["player", "salary", "team_rank"]]
            team_df["team_abbrev"] = team
            team_df["team_name"] = div[team]
            total_df = total_df.append(team_df)
            
total_df = total_df.reset_index(drop=True)
total_df
```

```python
# Exporting data as JSON file
total_df.to_json('2016_2017_salaries.json')
```

```python
players = player.PlayerList(only_current=0)
players = players.info()
players
```

```python
player_strings = [x.encode('utf-8') for x in players['DISPLAY_FIRST_LAST']]
player_strings = [x.lower() for x in player_strings]
player_strings
```

```python
player_ids = []

for player in total_df['player']:
    lowercase = player.lower()
    try:
        player_index = player_strings.index(lowercase)
        player_id = players.iloc[player_index, 0]
        player_ids.append(player_id)
    except:
        player_ids.append("Player Not Found")
    
player_ids
```

```python
player_series = pd.Series(player_ids)
total_df["player_id"] = player_series
total_df
```

```python
players_not_found = total_df.loc[total_df["player_id"] == "Player Not Found", :]
players_not_found
```
